```python
import pandas as pd
import re

def extract_table_data(text):
    # Step 1: Split text into lines
    lines = text.split('\n')

    # Step 2: Find the table starting point after 'Total Acquisition Estimates and Quantities'
    start_index = -1
    for i, line in enumerate(lines):
        if '(U) Total Acquisition Estimates and Quantities' in line:
            start_index = i
            break

    if start_index == -1:
        print("Table not found")
        return pd.DataFrame()

    # Step 3: Extract relevant table lines
    table_lines = lines[start_index+1:]

    # Step 4: Split each line by 2 or more spaces to handle table columns
    table_data = []
    for line in table_lines:
        split_line = re.split(r'\s{2,}', line.strip())
        # Only keep rows that have at least 2 columns (ignore headers and bad splits)
        if len(split_line) >= 2:
            table_data.append(split_line)

    # Step 5: Ensure all rows have the correct number of columns (at least 5), fill missing columns with None
    max_columns = 5  # We expect 5 columns
    cleaned_table_data = [row[:max_columns] + [None] * (max_columns - len(row)) for row in table_data]

    # Step 6: Create the DataFrame
    columns = ['Category', 'Objective', 'Threshold', 'Current Estimate', 'PB 2025']
    df = pd.DataFrame(cleaned_table_data, columns=columns)

    # Step 7: Filter out irrelevant rows during parsing
    #phases = ['RDT&E', 'Procurement', 'MILCON', 'O&M', 'R&MF']
    #df = df[df['Category'].isin(phases)]

    # Step 8: Convert numeric values to appropriate format
    for col in ['Objective', 'Threshold', 'Current Estimate', 'PB 2025']:
        df[col] = pd.to_numeric(df[col].str.replace(',', '').str.replace('*', ''), errors='coerce')
    
    return df

# Example usage
with open('parsed_output3.txt', 'r') as f:
    text = f.read()

df_result = extract_table_data(text)

# Display the DataFrame
print(df_result)
```
